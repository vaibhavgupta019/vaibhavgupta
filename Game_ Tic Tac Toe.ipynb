```
from IPython.display import clear_output                        #importing clear_output function

def display_board(board):    

    '''

     Function to display Tic Tac Toe board 

     Args: board
     Output:  | | 
             - - -
              | | 
             - - -
              | | 

    '''                                   
    clear_output()                                               # used to clear output console

    print(board[7] + "|" + board[8] + "|" + board[9])
    print("- - -")
    print(board[4] + "|" + board[5] + "|" + board[6])
    print("- - -")
    print(board[1] + "|" + board[2] + "|" + board[3])
```

```
def player_marker_choice():

    '''

    Function to take player's choice of marker from X or O

    Args: marker 
    Output: returns output in the form of tuple (player1_marker,player2_marker)

    '''
    marker = ""                                                                    #Intial value of marker is blank space or NULL
    
    while not (marker == "X" or marker == "O"):
        marker = input("{Player 1: Which one do you want to be ( X or O)").upper()

    if marker == "X":
        return ("X", "O")
    else:
        return ("O", "X")
```

```
def place_board_marker(board, marker, position):

    '''
    Function to place marker in the specified position index

    Args: board,marker,position
    output: the marker is placed on the specified board index
              7|8|9
              - - - 
              4|5|6
              - - - 
              1|2|3

    '''
    board[position] = marker
```

```
def win_check(board, mark):

    '''
    Function to check winner

    Args: board,mark
    Output: returns True or False based on the specified conditions.
            All possible combination are used ----> 3 rows, 3 columns, 2 diagonals
    '''

    return (
           (
            board[7] == mark and board[8] == mark and board[9] == mark        # across the top
        )         
        or (  
            board[4] == mark and board[5] == mark and board[6] == mark        # across the middle
        )
        or (  
            board[1] == mark and board[2] == mark and board[3] == mark        # across the bottom
        )
        or (  
            board[7] == mark and board[4] == mark and board[1] == mark        # down the left side
        )
        or ( 
            board[8] == mark and board[5] == mark and board[2] == mark        # down the middle
        )
        or (  
            board[9] == mark and board[6] == mark and board[3] == mark        # down the right side
        )
        or (  
            board[7] == mark and board[5] == mark and board[3] == mark        # diagonal     
        )
        or (
            board[9] == mark and board[5] == mark and board[1] == mark        # diagonal 
        )       
    )  
```

```
import random                                      # importing random library to use randint function

def choose_first():

    '''
    Function to choose which player will go first
    
    Args: 0 or 1
    Output: returns Player 1 or 2 randomly
    '''
    if random.randint(0, 1) == 0:
        return 'Player 2'
    else:
        return 'Player 1'
```

```
def space_check(board, position):
    '''
    Function to check for empty space for the  specified index position

    Args: board, position
    Output: returns true if empty else false
    '''
    return board[position] == ' '
```

```
def full_board_check(board):
    '''
    Function to check whether the board is full

    Args: board
    Output: Returns True or False based on checking the space_check() function for all indexes
    '''
    for i in range(1,10):
        if space_check(board, i):
            return False
    return True
```

```
def player_choice(board):
    '''
    Function to take players position choice

    Args: board
    Output: returns position index based on checking 2 conditions
            1. the position must be in [1,2,3,4,5,6,7,8,9]
            2. the position should be empty, for this space_check() function is used
    '''
    position = 0                                                                      #intial value of variable position
    
    while position not in range(1,10) or not space_check(board, position):
        position = int(input('Choose your next position: (1-9) '))
        
    return position
```

```
def replay():
    '''
    Function to check whether user wants to play the game again

    Input: Takes input and converts to lowercase using lower() function
    Output: returns true if the starts with 'y', checking using startswith() function
    '''
    return input('Do you want to play again? Enter Yes or No: ').lower().startswith('y')
```

```
'''
                                                      MAIN FUNCTION
-------------------------------------------------------------------------------------------------------------------------------
-------------------------------------------------------------------------------------------------------------------------------
Variable List 
----------------                                                    
player1_marker: To store the marker choice of Player 1
player2_marker: To store the marker choice of Player 2
turn: To store which player is going first
play_game: Used to intially start the game by checking if the player is ready or not
game_on: To start the game based on the value of play_game variable

'''
print('Welcome to Tic Tac Toe!')

while True:
    # Reset the board
    theBoard = [' '] * 10
    player1_marker, player2_marker = player_marker_choice()
    turn = choose_first()
    print(turn + ' will go first.')
    
    play_game = input('Are you ready to play? Enter Yes or No.')
    
    if play_game.lower()[0] == 'y':
        game_on = True
    else:
        game_on = False

    while game_on:
        if turn == 'Player 1':
            # Player1's turn.
            
            display_board(theBoard)
            position = player_choice(theBoard)
            place_board_marker(theBoard, player1_marker, position)

            if win_check(theBoard, player1_marker):
                display_board(theBoard)
                print('Congratulations! You have won the game!')
                game_on = False
            else:
                if full_board_check(theBoard):
                    display_board(theBoard)
                    print('The game is a draw!')
                    break
                else:
                    turn = 'Player 2'

        else:
            # Player2's turn.
            
            display_board(theBoard)
            position = player_choice(theBoard)
            place_board_marker(theBoard, player2_marker, position)

            if win_check(theBoard, player2_marker):
                display_board(theBoard)
                print('Player 2 has won!')
                game_on = False
            else:
                if full_board_check(theBoard):
                    display_board(theBoard)
                    print('The game is a draw!')
                    break
                else:
                    turn = 'Player 1'

    if not replay():
        break
```
